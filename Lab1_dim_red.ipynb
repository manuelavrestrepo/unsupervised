## 5. Train a naive logistic regression on raw MNIST images to distinguish between 0s and 8s. We are calling this our baseline. What can you tell about the baseline performance?

````python
# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Filter images labeled as 0s and 8s
indices_0s = (y_train == 0)
indices_8s = (y_train == 8)

X_train_0s = X_train[indices_0s]
X_train_8s = X_train[indices_8s]
y_train_0s = y_train[indices_0s]
y_train_8s = y_train[indices_8s]
````

````python
# Display an example image of the digit '8'
plt.imshow(X_train_8s[200], cmap='gray')  # Display the image
plt.title('Example of the digit "8"')  # Add a title to the plot
plt.axis('off')  # Turn off axis labels for better visualization
plt.show()

# Display an example image of the digit '0'
plt.imshow(X_train_0s[100], cmap='gray')  # Display the image
plt.title('Example of the digit "0"')  # Add a title to the plot
plt.axis('off')  # Turn off axis labels for better visualization
plt.show()
````

````python
# Combine images and labels for 0s and 8s
X_combined = np.concatenate((X_train_0s, X_train_8s), axis=0)
y_combined = np.concatenate((y_train_0s, y_train_8s), axis=0)

# Flatten the images and normalize pixel values
X_flattened = X_combined.reshape(X_combined.shape[0], -1)  # Flatten images
X_normalized = X_flattened / 255.0  # Normalize pixel values to [0, 1]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y_combined, test_size=0.2, random_state=42)

# Train a Naive Logistic Regression Model
logistic_regression_model = LogisticRegression(random_state=42)

# Train the model
logistic_regression_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logistic_regression_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Logistic Regression model: {accuracy:.2f}")
````

**Conclusion**
With an accuracy of 0.99 for the logistic regression model trained on raw MNIST images to distinguish between 0s and 8s, you can conclude that the baseline performance is exceptionally high. An accuracy of 0.99 indicates that the model correctly predicts the digit (either 0 or 8) for 99% of the instances in the dataset. The high accuracy suggests that the model is effective in distinguishing between the digits 0 and 8 based on the raw pixel values. ## 8. What strategies do you know (or can think of) in order to make PCA more robust? (Bonus points for implementing them)

### Strategies for Robust PCA in Unsupervised Learning

To make PCA in an unsupervised setting more robust, you can consider the following strategies:

### Data Preprocessing:

- **Scaling:** Standardize or normalize the data before applying PCA to ensure that all features contribute equally to the variance.
- **Handling Missing Values:** Address missing values appropriately, either by imputing them or removing the corresponding samples.

### Outlier Handling:

- **Outlier Detection:** Identify and handle outliers, as they can significantly impact the principal components. Robust methods for outlier detection, such as the use of median absolute deviation, can be considered.

### Choosing the Number of Components:

- **Variance Retention:** Instead of using an arbitrary number of components, choose the number of components that retain a significant amount of varianr of components.

### Handling Categorical Data:

- **Encoding Categorical Variables:** If your dataset includes categorical variables, encode them appropriately before applying PCA.

### Incremental PCA:

- **Batch Processing:** For large datasets, consider using Incremental PCA, which processes data in batches, making it more memory-efficient.

### Kernel PCA:

- **Non-Linearity:** If the data has a non-linear structure, Kernel PCA can be applied with a suitable kernel to capturers, such as Robust PCA algorithms.


````python
def preprocess_data(X, categorical_columns=None):
    """
    Preprocesses the input data for unsupervised learning using PCA.

    Parameters:
    - X: The input dataset.
    - categorical_columns: List of column indices corresponding to categorical features.

    Returns:
    - X_preprocessed: The preprocessed dataset.
    """

    # Step 1: Scaling (Standardization)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Step 2: Handling Missing Values
    imputer = SimpleImputer(strategy='mean')
    X_imputed = imputer.fit_transform(X_scaled)

    # Step 3: Handling Categorical Data (if specified)
    if categorical_columns is not None:
        encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
        X_categorical_encoded = encoder.fit_transform(X[:, categorical_columns])
        X_preprocessed = np.hstack((X_imputed, X_categorical_encoded))
    else:
        X_preprocessed = X_imputed

    return X_preprocessed
````

````python
# Assuming X_train_0s and X_train_8s are your flattened images for 0s and 8s
X_train_0s_flatten = X_train_0s.reshape(X_train_0s.shape[0], -1)
X_train_8s_flatten = X_train_8s.reshape(X_train_8s.shape[0], -1)

# Combine the data
X_combined = np.vstack((X_train_0s_flatten, X_train_8s_flatten))
y_combined = np.hstack((np.zeros(X_train_0s_flatten.shape[0]), np.ones(X_train_8s_flatten.shape[0])))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, y_combined, test_size=0.2, random_state=42)

# Data preprocessing
X_preprocessed = preprocess_data(X_train)

# Incremental PCA
pca_model = PCA.PCA(n_components=2)
X_pca = pca_model.fit_transform(X_preprocessed)

# Plot the results
plt.scatter(X_pca[y_train == 0, 0], X_pca[y_train == 0, 1], label='Digit 0', color='teal')
plt.scatter(X_pca[y_train == 1, 0], X_pca[y_train == 1, 1], label='Digit 8', color='gold')
plt.title('PCA with MNIST Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
````

````python
# Train Logistic Regression on Incremental PCA transformed data
model_incremental_pca = LogisticRegression()
model_incremental_pca.fit(X_pca, y_train)

# Transform the test data using the same Incremental PCA model
X_test_pca = pca_model.transform(X_test)

# Predictions on the test set
y_pred_incremental_pca = model_incremental_pca.predict(X_test_pca)

# Evaluate accuracy
accuracy_incremental_pca = accuracy_score(y_test, y_pred_incremental_pca)

print(f'Accuracy with Incremental PCA: {accuracy_incremental_pca:.2f}')
````

### 9. UMAP (Uniform Manifold Approximation and Projection):

**Mathematical Principles:**
UMAP is a dimensionality reduction technique that aims to preserve both local and global structures of high-dimensional data. The algorithm is based on the following principles:

- **Fuzzy Topological Structure:** UMAP constructs a fuzzy topological representation of the data, capturing both local and global relationships. It uses the concept of simplicial sets to create a weighted graph.

- **Low-Dimensional Embedding:** UMAP seeks to find a low-dimensional representation that preserves the fuzzy topological structure. It optimizes a cost function that balances the preservation of local neighborhoods and global structures.

- **Riemannian Geometry:** UMAP leverages Riemannian geometry to define a metric on the fuzzy topological representation. This metric is used to guide the optimization process for finding the low-dimensional embedding.

**Usefulness:**
UMAP is useful for various tasks, including:

- **Visualizing High-Dimensional Data:** UMAP is particularly effective for creating visually appealing and interpretable low-dimensional representations of complex high-dimensional data.

- **Clustering and Pattern Recognition:** UMAP can reveal intrinsic clusters and patterns in the data, making it valuable for tasks such as clustering and identifying groups of similar data points.

- **Preserving Local and Global Structures:** Unlike some other dimensionality reduction methods, UMAP excels at preserving both local and global structures in the data, providing a more comprehensive representation.

### 10. LDA (Latent Dirichlet Allocation):

**Mathematical Principles:**
LDA is a generative probabilistic model for a given set of documents. The underlying principles include:

- **Probabilistic Modeling:** LDA assumes that documents are mixtures of topics, and topics are mixtures of words. It models the generative process of how documents are created based on these mixtures.

- **Dirichlet Distributions:** LDA uses Dirichlet distributions to model the mixtures of topics within documents and the mixtures of words within topics. These distributions capture the uncertainty and variability in the assignment of words to topics.

- **Bayesian Inference:** LDA employs Bayesian inference to estimate the parameters of the model. It iteratively refines its estimates based on observed data, adjusting the mixtures of topics and words to best explain the given set of documents.

**Usefulness:**
LDA is widely used in natural language processing and text analysis tasks:

- **Topic Modeling:** LDA is specifically designed for topic modeling, where it identifies underlying topics in a collection of documents. Each document is represented as a distribution over topics, and each topic is represented as a distribution over words.

- **Document Classification:** LDA can be used for document classification by assigning topics to documents based on their topic distributions.

- **Understanding Document Collections:** LDA helps in understanding the thematic structure of large document collections, aiding in tasks such as information retrieval and content organization.
